```python
import cv2
import numpy as np
from matplotlib import pyplot as plt
image = cv2.imread('caballos.jpg')
```

```python
B, G, R = cv2.split(image)

# Creamos una matriz de ceros con las dimensiones de la imagen h x w.
zeros = np.zeros(image.shape[:2], dtype = "uint8")

row, col = 1, 3
fig, axs = plt.subplots(row, col, figsize=(15, 10))
fig.tight_layout()
 
axs[0].imshow(cv2.merge([R, zeros, zeros]) )
axs[0].set_title('Rojo')

axs[1].imshow(cv2.merge([zeros, G, zeros]))
axs[1].set_title('Verde')

axs[2].imshow(cv2.merge([zeros, zeros, B]))
axs[2].set_title('Azul')

plt.show()
```

```python
# Guardamos las imagenes
cv2.imwrite('rojo.jpg', cv2.merge([zeros, zeros, B]))
cv2.imwrite('verde.jpg', cv2.merge([zeros, G, zeros]))
cv2.imwrite('azul.jpg', cv2.merge([R, zeros, zeros]))
```
