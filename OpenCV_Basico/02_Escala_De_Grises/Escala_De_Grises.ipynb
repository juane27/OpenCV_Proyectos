<h1>Escala de grises<h1>
<h5>Vamos a convertir imagenes de color a imagenes en escalas de grises con OpenCV.<h5>


```python
import cv2
import numpy as np
from matplotlib import pyplot as plt
# Ver versi√≥n de OpenCV:
cv2.__version__
```

```python
# Leemos nuestra imagen
image = cv2.imread('caballos.jpg')

# Utilizamos cvtColor para convertir la imagen a escala de grises
# El primer par√°metro es la imagen a convertir y el segundo es el tipo de conversi√≥n
gray_image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

# Para comparar las imagenes vamos a crear una columna por imagen.
row, col = 1, 2
fig, axs = plt.subplots(row, col, figsize=(15, 10))
fig.tight_layout()
 
# Mostramos la imagen original en la columna 0
axs[0].imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
axs[0].set_title('A color (original)')

# Mostramos la imagen en escala de grises en la columna 1
axs[1].imshow(cv2.cvtColor(gray_image, cv2.COLOR_BGR2RGB))
axs[1].set_title('En escala de grises')

plt.show()
```

```python
# Para guardar la imagen en escala de grises
cv2.imwrite('Caballos_escala_de_grises.jpg', gray_image)
```
