<h2>Dibujando imagenes y formas con OpenCV</h2>
Empezamos haciendo un cuadrado negro:

```python
import cv2
import numpy as np
from matplotlib import pyplot as plt
```

```python
# Crear cuadrado negro.(imagen con 3 canales)
image = np.zeros((480,640,3), np.uint8)

# (Imagen en blanco y negro)
image_bw = np.zeros((480,640), np.uint8)

row, col = 1, 2
fig, axs = plt.subplots(row, col, figsize=(15, 10))
fig.tight_layout()
 
axs[0].imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
axs[0].set_title('Original RGB Colors')

axs[1].imshow(cv2.cvtColor(image_bw, cv2.COLOR_BGR2RGB))
axs[1].set_title('Original RGB Colors')

plt.show()
```

<h3>Dibujamos una linea azul<h3>

```python
# Dibujamos una linea horizontal azul en el medio de la imagen de un ancho de 5 pixels
# 1° valor coordenadas de origen, 2° valor coordenadas de destino, 3° color de la linea, 4° grosor de la linea
image = np.zeros((512,512,3), np.uint8)
cv2.line(image, (0,220), (500,220), (260,0,0), 5)
 
plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
plt.title('Linea azul')

plt.show()
```

<h3>Dibujamos un rectangulo<h3>

```python
# Dibujamos un rectangulo violeta
image = np.zeros((512,512,3), np.uint8)

cv2.rectangle(image, (100,100), (400,250), (255,50,127), 20)
plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))

plt.title('Rectangulo violeta')

plt.show()
```

<h3>Dibujamos un circulo<h3>

```python
image = np.zeros((512,512,3), np.uint8)

cv2.circle(image, (150, 350), 100, (125,75,50), 50) 

plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
plt.title('Circulo')
plt.show()
```

<h3>Dibujamos poligono<h3>

```python
image = np.zeros((512,512,3), np.uint8)

# Definimos los puntos del poligono
pts = np.array( [[40,50], [400,50], [300,400], [50,500]], np.int32)

# reshape a los datos
pts = pts.reshape((-1,1,2))

cv2.polylines(image, [pts], True, (0,45,255), 30)

plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
plt.title('Poligono')
plt.show()
```

cv2.putText(image, 'texto a mostrar', punto de inicio abajo a la izquierda, Fuente, Tamaño, Color, Grosor)

```python
image = np.zeros((512,512,3), np.uint8)

cv2.putText(image, 'Caballos', (200,290), cv2.FONT_HERSHEY_COMPLEX, 1, (100,170,0), 3)  
plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
plt.title('Texto')
plt.show()
```
