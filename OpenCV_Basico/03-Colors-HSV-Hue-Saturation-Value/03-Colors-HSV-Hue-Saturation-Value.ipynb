<h1>Colors, HSV, Hue, Saturation, Value</h1>
<h5>¿Qué significa HSV?</h5>
El modelo HSV (del inglés Hue, Saturation, Value – Matiz, Saturación, Valor), define un modelo de color en términos de sus componentes.
Los valores de los componentes comprenden los siguientes rangos:
#H: 0 - 180, S: 0 - 255, V: 0 - 255

```python
import cv2
import numpy as np
from matplotlib import pyplot as plt

# Leemos nuestra imagen
image = cv2.imread('caballos.jpg')

hsv_image = cv2.cvtColor(image, cv2.COLOR_BGR2HSV)


row, col = 2, 2
fig, axs = plt.subplots(row, col, figsize=(15, 10))
fig.tight_layout()
 
axs[0][0].imshow(cv2.cvtColor(hsv_image, cv2.COLOR_BGR2RGB))
axs[0][0].set_title('HSV')

axs[0][1].imshow(cv2.cvtColor(hsv_image[:, :, 0], cv2.COLOR_BGR2RGB))
axs[0][1].set_title('Canal Hue')

axs[1][0].imshow(cv2.cvtColor(hsv_image[:, :, 1], cv2.COLOR_BGR2RGB))
axs[1][0].set_title('Canal Saturation')

axs[1][1].imshow(cv2.cvtColor(hsv_image[:, :, 2], cv2.COLOR_BGR2RGB))
axs[1][1].set_title('Canal Value')

plt.show()
```

```python
# Guardamos las imagenes
cv2.imwrite('hsv_image.jpg', hsv_image)
cv2.imwrite('hue_image.jpg', hsv_image[:, :, 0])
cv2.imwrite('saturation_image.jpg', hsv_image[:, :, 1])
cv2.imwrite('value_image.jpg', hsv_image[:, :, 2])
```
