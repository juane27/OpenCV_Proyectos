<h1>Edge Detection & Image Gradients</h1>   


```python
import cv2
import numpy as np
from matplotlib import pyplot as plt
    
```

```python
# Extremos Bordes Sobel 
image = cv2.imread('torito.jfif')
sobel_x = cv2.Sobel(image, cv2.CV_64F, 0, 1, ksize=5)
sobel_y = cv2.Sobel(image, cv2.CV_64F, 1, 0, ksize=5)
```

```python
sobel_OR = cv2.bitwise_or(sobel_x, sobel_y)
laplacian = cv2.Laplacian(image, cv2.CV_64F)
```

```python
row, col = 1, 2
fig, axs = plt.subplots(row, col, figsize=(15, 10))
fig.tight_layout()
 
axs[0].imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
axs[0].set_title('Original Image')
cv2.imwrite('original_image.png', image)

# Debemos proveer dos valores: threshold1 y threshold2. Cualquier gradiente 
# mayor que threshold2 es considerado un borde. Cualquier valor menor que  
# threshold1 es cosiderado como NO borde. Valores entre threshold1 y 
# threshold2 pueden ser clasificados como bordes o NO segun como se conecten.

canny = cv2.Canny(image, 150, 260)
axs[1].imshow(cv2.cvtColor(canny, cv2.COLOR_BGR2RGB))
axs[1].set_title('canny')
cv2.imwrite('canny.png', canny)

plt.show()
```
