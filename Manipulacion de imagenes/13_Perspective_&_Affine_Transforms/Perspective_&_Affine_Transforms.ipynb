<h1>Getting Perpsective Transform</h1>
Perspectiva

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

image = cv2.imread('apuntes.jpeg')
```

```python
row, col = 1, 2
fig, axs = plt.subplots(row, col, figsize=(15, 10))
fig.tight_layout()

axs[0].imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
axs[0].set_title('Original RGB Colors')
cv2.imwrite('original_rgb.jpg', image)


# Cordenadas de las 4 esquinas de la imagen orginal
points_A = np.float32([[255,700], [660,650], [330,1360], [900,1200]])

# Cordenadas de las 4 esquinas del output deseado
# Ratio de papel A4 1 : 1.41
points_B = np.float32([[0,0], [420,0], [0,594], [420,594]])
# Usamos los dos sets de puntos para calcular la matriz de transformacion de perspectiva M    
M = cv2.getPerspectiveTransform(points_A, points_B)
 
warped = cv2.warpPerspective(image, M, (420,594))

axs[1].imshow(cv2.cvtColor(warped, cv2.COLOR_BGR2RGB))
axs[1].set_title('WarpPerspective')
cv2.imwrite('warped.jpg', warped)

plt.show()
```
