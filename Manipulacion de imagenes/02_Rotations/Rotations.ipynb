<h1>Rotations</h1>
cv2.getRotationMatrix2D(rotation_center_x, rotation_center_y, angle of rotation, scale)

```python
import cv2
import numpy as np
from matplotlib import pyplot as plt
```

```python
image = cv2.imread('toro.jpg')
height, width = image.shape[:2]

# Dividimos por dos para rotar la imagen sobre su centro.
rotation_matrix = cv2.getRotationMatrix2D((width/2, height/2), 90, .5)

rotated_image = cv2.warpAffine(image, rotation_matrix, (width, height))
```

```python
plt.imshow(cv2.cvtColor(rotated_image, cv2.COLOR_BGR2RGB))
plt.title('Rotated Image')

plt.show()
```

<h3>Como eliminamos todo el espacio negro sobrante?</h3>

Podemos hacerlo en rotaciones con la funcin cv2.transpose function

```python
#Otra forma de rotar
img = cv2.imread('toro.jpg')

rotated_image = cv2.transpose(img)

plt.imshow(cv2.cvtColor(rotated_image, cv2.COLOR_BGR2RGB))
plt.title('Rotated Image - Method 2')

plt.show()
```

```python
# Para hacer un giro horizontal:
flipped = cv2.flip(image, 1)

plt.imshow(cv2.cvtColor(flipped, cv2.COLOR_BGR2RGB))
plt.title('Horizontal Flip')

plt.show()
```
