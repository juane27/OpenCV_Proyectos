```python
import cv2
import numpy as np
from matplotlib import pyplot as plt
image = cv2.imread('toro.jpg')
```

```python
image.shape
```

```python
x, y = image.shape[:2]
```

```python
height, width = image.shape[:2]

# Cordenadas de inicio (Arriba a la izquierda del rectangulo)
start_row, start_col = int(1500), int(1000)

# Cordenadas finales (Abajo a la derecha)
end_row, end_col = int(3600), int(4450)

# Usar indexing para cortar el rectangulo que deseamos.
cropped = image[start_row:end_row , start_col:end_col]
```

```python
row, col = 1, 2
fig, axs = plt.subplots(row, col, figsize=(15, 10))
fig.tight_layout()
 
axs[0].imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
axs[0].set_title('Original Image')
cv2.imwrite('original_image.png', image)

axs[1].imshow(cv2.cvtColor(cropped, cv2.COLOR_BGR2RGB))
axs[1].set_title('Cropped Image')
cv2.imwrite('cropped_image.png', cropped)

plt.show()
```
