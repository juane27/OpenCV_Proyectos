<h1>Translations</h1>
Esta es una transformación affine(que preserva lines y paralelismos, pero no necesariamente distancias y angulos) que simplemente cambia la posición de la imagen.

Utilizamos cv2.warpAffine para implementar esta transformación.

```python
import cv2
import numpy as np
from matplotlib import pyplot as plt
```

```python
image = cv2.imread('toro.jpg')

# Almacenamos la altura y ancho de la imagen.
height, width = image.shape[:2]

# 1/4 de altura y ancho
quarter_height, quarter_width = height/4, width/4


print ("height.shape ", height)
print ("quarter_height.shape ", quarter_height)
print ("width.shape ", height)
print ("quarter_width.shape ", quarter_width)
```

```python
#      | 1 0 Tx |
# T  = | 0 1 Ty |
# T es la matriz de nuestra transformacion
T = np.float32([[1, 0, quarter_width], 
                [0, 1,quarter_height]])

# Imprimimos T
print (T)
```

```python
# Utilizamos warpAffine para transformar la imagen usando la matriz T
img_translation = cv2.warpAffine(image, T, (width, height))
print (img_translation.shape)
```

```python
plt.imshow(cv2.cvtColor(img_translation, cv2.COLOR_BGR2RGB))
plt.title('Translation')

plt.show()
```
