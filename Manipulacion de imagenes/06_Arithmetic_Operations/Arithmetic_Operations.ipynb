<h1>Arithmetic Operations</h1>
Hay operaciones simples que nos permiten directamente añadir o sustraer a la intensidad del color.

Calcula la operación por elemento de dos matrices. El resultado del efecto es el aumento o la disminución del brillo.

```python
import cv2
import numpy as np
from matplotlib import pyplot as plt
image = cv2.imread('toro.jpg')
```

```python
# Crear una matriz de unos
# Esto nos da una matriz de 100 en todos los pixeles
M = np.ones(image.shape, dtype = "uint8") * 100 
```

```python
row, col = 1, 3
fig, axs = plt.subplots(row, col, figsize=(15, 10))
fig.tight_layout()

axs[0].imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
axs[0].set_title('Original Image')
cv2.imwrite('image.png', image)
 
# Utilizamos esto para añadir la matriz M a nuestra imagen.
# Efecto de aumento de brillo
added = cv2.add(image, M)
axs[1].imshow(cv2.cvtColor(added, cv2.COLOR_BGR2RGB))
axs[1].set_title('Added Image')
cv2.imwrite('Added_image.png', added)

# Sustraemos la matriz M de nuestra imagen.
# Efecto de disminucion de brillo
subtracted = cv2.subtract(image, M)
axs[2].imshow(cv2.cvtColor(subtracted, cv2.COLOR_BGR2RGB))
axs[2].set_title('Subtracted')
cv2.imwrite('subtracted_image.png', subtracted)

plt.show()
```
