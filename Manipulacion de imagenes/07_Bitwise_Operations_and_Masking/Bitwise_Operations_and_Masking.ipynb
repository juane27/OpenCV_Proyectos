<h1>Bitwise Operations and Masking</h1>
Para demostrar estas operaciones vamos a hacer algunas im√°genes

```python
import cv2
import numpy as np
from matplotlib import pyplot as plt
image = cv2.imread('toro.jpg')
```

```python
row, col = 1, 2
fig, axs = plt.subplots(row, col, figsize=(10, 5))
fig.tight_layout()

# 2 dimensiones por que es en escala de grises.
# Si queremos trabajar una imagen con color usamos 3 dimensiones.
# rectangle = np.zeros((300, 300, 3),np.uint8)

# Haciendo un cuadrado
square = np.zeros((300, 300), np.uint8)
cv2.rectangle(square, (50, 50), (250, 250), 255, -2)
axs[0].imshow(cv2.cvtColor(square, cv2.COLOR_BGR2RGB))
axs[0].set_title('Square Image')
cv2.imwrite('square.png', square)
 
    
# Haciendo un elipse
ellipse = np.zeros((300, 300), np.uint8)
cv2.ellipse(ellipse, (150, 150), (150, 150), 30, 0, 180, 255, -1)
axs[1].imshow(cv2.cvtColor(ellipse, cv2.COLOR_BGR2RGB))
axs[1].set_title('Ellipse Image')
cv2.imwrite('ellipse_image.png', ellipse)

plt.show()
```

<h2>Experimentando con algunas operaciones bitwise</h2>

```python
row, col = 2, 2
fig, axs = plt.subplots(row, col, figsize=(10, 5))
fig.tight_layout()

# Muestra unicamente donde se interceptan
And = cv2.bitwise_and(square, ellipse)
axs[0][0].imshow(cv2.cvtColor(And, cv2.COLOR_BGR2RGB))
axs[0][0].set_title('And Image')
cv2.imwrite('and.png', And)

# Muestra donde esta el cuadrado o el elipse
bitwiseOr = cv2.bitwise_or(square, ellipse)
axs[0][1].imshow(cv2.cvtColor(bitwiseOr, cv2.COLOR_BGR2RGB))
axs[0][1].set_title('bitwiseOr Image')
cv2.imwrite('bitwiseOr.png', bitwiseOr)

# Muestra donde existe
bitwiseXor = cv2.bitwise_xor(square, ellipse)
axs[1][0].imshow(cv2.cvtColor(bitwiseXor, cv2.COLOR_BGR2RGB))
axs[1][0].set_title('bitwiseXor Image')
cv2.imwrite('bitwiseXor.png', bitwiseXor)

# Muestra todo lo que no es parte del cuadrado
bitwiseNot_sq = cv2.bitwise_not(square)
axs[1][1].imshow(cv2.cvtColor(bitwiseNot_sq, cv2.COLOR_BGR2RGB))
axs[1][1].set_title('bitwiseNot_sq Image')
cv2.imwrite('bitwiseNot_sq.png', bitwiseNot_sq)

plt.show()
```
