<h1>Encontrando la vaca negra</h1>

```python
import cv2
import numpy as np
from matplotlib import pyplot as plt
```

```python

image_template = cv2.imread('drone-cows.jpg')
gray_template = cv2.cvtColor(image_template, cv2.COLOR_BGR2GRAY)

row, col = 1, 2
fig, axs = plt.subplots(row, col, figsize=(15, 10))
fig.tight_layout()
 
axs[0].imshow(cv2.cvtColor(image_template, cv2.COLOR_BGR2RGB))
axs[0].set_title('¿Dónde esta la vaca negra?')
cv2.imwrite('black-cow.jpg', image_template)

axs[1].imshow(cv2.cvtColor(gray_template, cv2.COLOR_BGR2RGB))
axs[1].set_title('Imagen gris')
cv2.imwrite('gray_black-cow.jpg', gray_template)

plt.show()
```

```python
# Cargar template
gray_cow = cv2.imread('template.jpg',0)
color_cow = cv2.imread('template.jpg')

row, col = 1, 2
fig, axs = plt.subplots(row, col, figsize=(5, 3))
fig.tight_layout()
 
axs[0].imshow(cv2.cvtColor(gray_cow, cv2.COLOR_BGR2RGB))
axs[0].set_title('Black cow template')
cv2.imwrite('cow_template.jpg', gray_cow)

axs[1].imshow(cv2.cvtColor(color_cow, cv2.COLOR_BGR2RGB))
axs[1].set_title('Black cow color template')
cv2.imwrite('cow_color_template.jpg', color_cow)

plt.show()
```

```python
result = cv2.matchTemplate(image_template, color_cow, cv2.TM_CCOEFF)
min_val, max_val, min_loc, max_loc = cv2.minMaxLoc(result)
print (len(result))

#Crear caja con limites
top_left = max_loc
bottom_right = (top_left[0] + 220, top_left[1] + 130)
cv2.rectangle(image_template, top_left, bottom_right, (0,255,0), 5)

plt.imshow(cv2.cvtColor(image_template, cv2.COLOR_BGR2RGB))
plt.title('¿Dónde esta la vaca negra?'); plt.show()
cv2.imwrite('output_template.jpg', image_template)
```
