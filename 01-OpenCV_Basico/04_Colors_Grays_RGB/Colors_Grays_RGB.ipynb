```python
import cv2
import numpy as np
from matplotlib import pyplot as plt
image = cv2.imread('caballos.jpg')
```

Vamos a analizar los colores individuales del primer pixel

```python
# Valores BGR para el primer pixel de la imagen 0,0
B, G, R = image[0, 0] 
print ('Valor de azul: ' + str(B), ' Valor de verde: ' + str(G), ' Valor de rojo: ' + str(R))
# Tama√±o de la imagen y canales
print (image.shape)
```

Vamos a ver que sucede si a la imagen la convertimos a escala de grises

```python
gray_img = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
print (gray_img[0, 0])
print (gray_img.shape)
```

Vemos que la imagen una vez transformada a escala de grises tiene solo 2 dimensiones. Cada pixel tiene un solo valor en el rango de 0 a 255.

<h3>Vamos a observar los canales individuales RGB<h3>


```python
# La funcion 'split' de OpenCV divide la imagen en 3 canales
B, G, R = cv2.split(image)

print (B.shape)

row, col = 1, 3
fig, axs = plt.subplots(row, col, figsize=(15, 10))
fig.tight_layout()
 
axs[0].imshow(cv2.cvtColor(R, cv2.COLOR_BGR2RGB))
axs[0].set_title('Red')

axs[1].imshow(cv2.cvtColor(G, cv2.COLOR_BGR2RGB))
axs[1].set_title('Green')

axs[2].imshow(cv2.cvtColor(B, cv2.COLOR_BGR2RGB))
axs[2].set_title('Blue')

plt.show()
```

```python
row, col = 1, 2
fig, axs = plt.subplots(row, col, figsize=(15, 10))
fig.tight_layout()
 
merged = cv2.merge([B, G, R]) 
axs[0].imshow(cv2.cvtColor(merged, cv2.COLOR_BGR2RGB))
axs[0].set_title('Uniendo los canales')

merged = cv2.merge([B+100, G, R])
axs[1].imshow(cv2.cvtColor(merged, cv2.COLOR_BGR2RGB))
axs[1].set_title('Canales unidos pero con un valor mayor de azul')

plt.show()
```

```python
# Para guardar la imagen:
cv2.imwrite('canales_unidos_mas_azul.jpg', merged)
```
