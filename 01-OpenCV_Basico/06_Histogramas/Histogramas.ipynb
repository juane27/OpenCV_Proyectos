<h2>Los histogramas son utiles para visualizar los componentes individuales de color</h2>
 Los histogramas capturan distribución de frecuencias de un conjunto de datos.


```python
import cv2
import numpy as np

# Necesitamos matplotlib para plotear nuestros histogramas
from matplotlib import pyplot as plt
```

**cv2.calcHist(images, channels, mask, histSize, ranges[, hist[, accumulate]])**

* images : Es la ruta de la imagen de tipo uint8 o float32. Tiene que expresarse entre corchetes, Ejemplo: "[img]".
* channels : También se deben expresar entre corchetes. Es el indice de canal para el cual calculamos el histograma. Por ejemplo, si nuestra imagen esta en escala de grisis, su valor es [0]. Si es una imagen de color podemos pasar [0], [1] or [2] para calcular el histograma de azul, verde o rojo respectivamente.
* mask : Mascara de imagen. Para calcular el histograma de una imagen completa se le da como valor "None".  Pero si queres analizar y calcular el histograma de cierta parte de la imagen, tenes que crear una imagen mascara y pasarla como mascara.
* histSize : Esto representa el BIN count. Se pasa entre corchetes, para una escala completa pasamos [256].
* ranges :Este es nuestro rango, usualmente: [0,256].

```python
image = cv2.imread('caballos.jpg')
histogram = cv2.calcHist([image], [0], None, [256], [0, 256])
```

numpy.ravel(a, order='C') --> Haciendo la imagen PLANA(flat)

* a (array_like): Input array. The elements in a are read in the order specified by order, and packed as a 1-D array.

* order {‘C’,’F’, ‘A’, ‘K’}, optional: The elements of a are read using this index order. ‘C’ means to index the elements in row-major, C-style order, with the last axis index changing fastest, back to the first axis index changing slowest. ‘F’ means to index the elements in column-major, Fortran-style order, with the first index changing fastest, and the last index changing slowest. Note that the ‘C’ and ‘F’ options take no account of the memory layout of the underlying array, and only refer to the order of axis indexing. ‘A’ means to read the elements in Fortran-like index order if a is Fortran contiguous in memory, C-like order otherwise. ‘K’ means to read the elements in the order they occur in memory, except for reversing the data when strides are negative. By default, ‘C’ index order is used.

```python
# Ploteamos nuestro histograma, ravel() aplana la imagen.
plt.hist(image.ravel(), 256, [0, 256])
plt.show()
```

```python
# Viendo los distintos canales de color
color = ('b', 'g', 'r')

# Separamos los colores y ploteamos cada uno
for i, col in enumerate(color):
    histogram2 = cv2.calcHist([image], [i], None, [256], [0, 256])
    plt.plot(histogram2, color = col)
    plt.xlim([0,256])
    
plt.show()
```
